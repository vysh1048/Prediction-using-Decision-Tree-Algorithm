```python
import sklearn.datasets as dataset
frame=dataset.load_iris()
df=pd.DataFrame(frame.data, columns=frame.feature_names)
df.head()
```

```python
x=frame.data
y=frame.target
print(y)
```

#### Decision Tree Algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
(x_train, x_test, y_train, y_test) = train_test_split(x, y, train_size=0.8, random_state=1)
```

Accuracy Score 

```python
algo = DecisionTreeClassifier()
algo.fit(x_train, y_train)
algo.score(x_test, y_test)
```

#### Visualizing the Decision Tree

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(algo, filled=True)
```

#### Testing

```python
pred=algo.predict(x_test)
pred
```

```python
df=pd.DataFrame({'Predicted':pred, 'Actual':y_test})
df['label']=df['Predicted'].replace(dict(enumerate(frame.target_names)))
df.head(10)
```
